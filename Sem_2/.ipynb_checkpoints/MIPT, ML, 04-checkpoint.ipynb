## Helper functions

```python
def get_learning_curve(estimator, object_ids, X_train=X_train, Y_train=Y_train,
                       X_test=X_test, Y_test=Y_test, min_object_size=10, step_size=1):
    metrics = defaultdict(list)
    sizes = range(min_object_size, len(object_ids) + step_size - 1, step_size)
    for size in sizes:
        # print 'processing size: %d' % size
        estimator.fit(X_train[object_ids[:size]], Y_train[object_ids[:size]])
        metrics['test_error'].append(1.0- accuracy_score(Y_test, estimator.predict(X_test)))
        metrics['train_error'].append(1.0- accuracy_score(Y_train, estimator.predict(X_train)))
    df = pd.DataFrame(metrics, index=sizes)
    return df
```

## STOLP

```python
def estimator_margins(estimator, X, Y):
    probabilities = estimator.predict_proba(X)
    margins = []
    for class_id, probability in zip(Y, probabilities):
        margin = sum(prob for prob_class, prob in zip(estimator.classes_, probability) if prob_class == class_id)
        if sum(probability) != margin:
            margin -= max(prob for prob_class, prob in zip(estimator.classes_, probability) if prob_class != class_id)
        margins.append(margin)
    return margins
```

```python
def STOLP(estimator, X_train, Y_train, outlier_threshold, error_rate,
          max_iteration_count=10, verbose=True, X_test=None, Y_test=None,
          outlier_heuristic=False, etalon_iteration_count=1
          ):
    metrics = defaultdict(list)
    good_indices = range(X_train.shape[0])
    
    # delete outliers
    estimator.fit(X_train, Y_train)
    classes = estimator.classes_
    margins = collect_margins(
        n_neighbors=estimator.n_neighbors,
        margin_calculator=calculate_weighted_margin if estimator.weights == 'distance' else calculate_uniform_margin,
        X_train=X_train[good_indices],
        Y_train=Y_train[good_indices],
    )
    good_indices = list(
        index
        for index, margin in enumerate(margins)
        if margin >= outlier_threshold
            and index in good_indices
    )
    
    # Find one etalon object for each class
    etalon_set = []
    for etalon_iteration in range(etalon_iteration_count):
        for class_id in classes:
            biggest_margin_id = sorted((
                    (margin, object_id)
                    for object_id, margin in enumerate(margins)
                    if Y_train[object_id] == class_id
                        and object_id in good_indices
                ), reverse=True)[0][1]
            good_indices.remove(biggest_margin_id)
            etalon_set.append(biggest_margin_id)
    
    # While not satisfied by quality or time constraints
    # add one new object to etalon set per each iteration
    iteration_count = 0
    current_error_rate = 1.01
    while (good_indices
           and iteration_count < max_iteration_count
           and current_error_rate >= error_rate
           ):
        estimator.fit(X_train[etalon_set], Y_train[etalon_set])
        margins = estimator_margins(estimator, X_train[good_indices], Y_train[good_indices])
        if outlier_heuristic:
            min_object_id = sorted(
                (margin, object_id)
                for object_id, margin in enumerate(margins)
                if margin >= outlier_threshold
            )[0][1]
        else:
            min_object_id = sorted((margin, object_id) for object_id, margin in enumerate(margins))[0][1]
        min_object_id = good_indices[min_object_id]
        good_indices.remove(min_object_id)
        etalon_set.append(min_object_id)
        
        current_error_rate = 1.0 * sum(1 for margin in margins if margin <= 0.0) / X_train.shape[0]
        iteration_count += 1
        
        # learning curves
        metrics['error_rate'].append(current_error_rate)
        metrics['min_margin'].append(min(margins))
        metrics['max_margin'].append(max(margins))
        metrics['avg_margin'].append(sum(margins) * 1.0 / len(margins))
        if X_test is not None and Y_test is not None:
            prediction = estimator.fit(X_train[etalon_set], Y_train[etalon_set]).predict(X_test)
            metrics['test_error'].append(1.0- accuracy_score(Y_test, prediction))
        if verbose:
            if iteration_count % 10 == 0:
                print iteration_count, current_error_rate, min(margins)
        
    return etalon_set, metrics
```

```python
def compare_teams(teams, first_team, second_team):
    merged = teams[first_team].merge(
        teams[second_team], left_index=True, right_index=True,
        suffixes=('_%s' % first_team, '_%s' % second_team)
    )
    merged.plot()
    return merged
```

## Feature Selection

```python
def add_one(estimator, feature_set, X_train, Y_train, X_test, Y_test, look_forward=10):
    Q_min = {0: float('+inf')}
    Q_min_set = {0: []}
    Q_min_index = 0
    for feature_set_size in range(1, len(feature_set) + 1):
        Q_min[feature_set_size] = float('+inf')
        for feature in set(feature_set).difference(set(Q_min_set[feature_set_size - 1])):
            estimator.fit(X_train[:, Q_min_set[feature_set_size - 1] + [feature]], Y_train)
            error = 1 - accuracy_score(Y_test, estimator.predict(X_test[:, Q_min_set[feature_set_size - 1] + [feature]]))
            if error < Q_min[feature_set_size]:
                Q_min_set[feature_set_size] = Q_min_set[feature_set_size - 1] + [feature]
                Q_min[feature_set_size] = error
                if Q_min[feature_set_size] < Q_min[Q_min_index]:
                    Q_min_index = feature_set_size
        print 'Q_min: %.4lf, set_size: %d, added: %s' % (Q_min[feature_set_size], feature_set_size, Q_min_set[feature_set_size][-1])
        if Q_min_index + look_forward <= feature_set_size:
            break
    return Q_min_set[Q_min_index]
```

```python
%time feature_set = add_one(estimator, range(X_train.shape[1]), X_train[:500], Y_train[:500], X_test[:500], Y_test[:500], look_forward=4)
%time 1 - accuracy_score(Y_test[:500], estimator.fit(X_train[:500, feature_set], Y_train[:500]).predict(X_test[:500, feature_set]))
# 35x speed-up
```
