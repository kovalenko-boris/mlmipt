# Imports

```python
from matplotlib import pylab
from sklearn.svm import SVC
from sklearn.datasets import load_digits
```

# Data

```python
digits = load_digits()
```

# Evaluation

```python
example_id = 10
#pylab.hot()
#pylab.jet()
pylab.gray()
#pylab.winter()
pylab.matshow(digits.images[example_id])
pylab.show()

print digits.target[example_id]
pylab.imshow(digits.images[example_id])
```

```python
classifier = SVC(C=100.0, gamma=0.001)
```

```python
classifier.fit(digits.data[:-1], digits.target[:-1])
classifier.predict(digits.data[-1]), digits.target[-1]
```

```python
classifier.fit(digits.data[:-10], digits.target[:-10])
classifier.predict(digits.data[-10:]), digits.target[-10:]
```
